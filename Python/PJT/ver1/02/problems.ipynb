```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# problemA
```

```python
# CSV 파일 경로
csv_path = "nlfx_data/NFLX.csv"

# CSV 파일 읽어오기 (첫 번째, 마지막 열 제외)
df = pd.read_csv(csv_path, usecols=range(0, 5))

# DataFrame 출력
df
```

```python
df.dtypes
```

```python
# problemB
```

```python
df['Date'] = pd.to_datetime(df['Date'])
after_2021_df = df.loc[df['Date'] >= '2021-01-01'].reset_index()
after_2021_df
```

```python
import matplotlib.pyplot as plt

x = after_2021_df['Date']
y = after_2021_df['Close']

# # 그래프 그리기
plt.plot(x, y, label='Close')

# 범례 추가
plt.legend()

# 그래프에 제목과 축 레이블 추가
plt.title('NLFX')
plt.xlabel('Date')
plt.ylabel('Close')

# x 축 회전
plt.xticks(rotation=45)

# 그래프 표시
plt.show()
```

```python
mean_df = after_2021_df.groupby(pd.Grouper(key = 'Date', freq = 'M'))['Close'].mean().reset_index()
```

```python
mean_df
```

```python
import matplotlib.pyplot as plt

x = mean_df['Date']
y4 = mean_df['Close']

# # 그래프 그리기
plt.plot(x, y4, label='Close')

# 범례 추가
plt.legend()

# 그래프에 제목과 축 레이블 추가
plt.title('NLFX')
plt.xlabel('Date')
plt.ylabel('Close')

# x 축 회전
plt.xticks(rotation=45)

# 그래프 표시
plt.show()
```

```python
after_2022_df = df.loc[df['Date'] >= '2022-01-01'].reset_index()

import matplotlib.pyplot as plt

x = after_2022_df['Date']
y1 = after_2022_df['Close']
y2 = after_2022_df['High']
y3 = after_2022_df['Low']

# # 그래프 그리기
plt.plot(x, y1, label='Close')
plt.plot(x, y2, label='High')
plt.plot(x, y3, label='Low')

# 범례 추가
plt.legend()

# 그래프에 제목과 축 레이블 추가
plt.title('NLFX')
plt.xlabel('Date')
plt.ylabel('Data')

# x 축 회전
plt.xticks(rotation=45)

# 그래프 표시
plt.show()
```
